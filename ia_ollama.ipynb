```python
def nuassist_chat():
    print("\n" + "="*50)
    print("      🏦 BEM-VINDO AO CHAT DO NUASSIST")
    print("="*50)

    # Função interna para exibir o menu de ajuda
    def exibir_ajuda():
        print("\nDúvidas Frequentes: [Saldo] [Rendimento] [Cartão] [Segurança]")
        print("Dica: Digite sua dúvida abaixo ou 'sair' para encerrar.")
        print("-" * 50)

    exibir_ajuda()

    while True:
        # Entrada livre do usuário
        user_input = input("\nVocê (pergunta livre): ")

        if user_input.lower() in ['sair', 'exit', 'parar']:
            print("\nNuAssist: Entendido. Tenha um ótimo dia!")
            break

        if not user_input.strip():
            continue

        print("NuAssist está digitando...")

        # Chamada para o Ollama
        try:
            # Tratamento básico para evitar erros de aspas no comando shell
            safe_input = user_input.replace("'", "").replace('"', "")

            # Executa o modelo e captura a resposta
            resposta_bruta = os.popen(f"ollama run nuassist '{safe_input}'").read().strip()

            # --- PARTE MODIFICADA: FILTRO DE TEMA ---
            if "COMANDO_INVALIDO" in resposta_bruta:
                print("\n" + "!"*50)
                print("❌ NuAssist: Sinto muito, mas essa pergunta é inválida.")
                print("Eu só posso ajudar com assuntos relacionados ao Banco!")
                print("!"*50)
                exibir_ajuda() # Mostra o menu novamente para orientar o usuário
            else:
                print(f"\nNuAssist: {resposta_bruta}")
                print("-" * 30)
            # ----------------------------------------

        except Exception as e:
            print(f"\nErro na conexão: {e}")

# Inicia o Chat
nuassist_chat()
```
